# Quest√£o 4
Gr√°fico dos desvios em fun√ß√£o da massa

```python
c = 299792458
G = 6.6743*10**(-11)
```

```python
class weak:
    def __init__(self, M, R):
        self.M = M
        self.R = R
        
    def angDesv(self):
        return 4*G*self.M/(c**2 * self.R)
        
```

```python
m_lua = 7.35*10**(22)
m_jup = 1.898*10**(27)
m_sol = 1.989*10**(30)
m_ana = 1.989*0.6*10**(30)
m_neu = 1.989*1.5*10**(30)
m_bh = 50*1.989*10**(30)

r_lua = 1.7374*10**(6)
r_jup = 6.9911*10**(7)
r_sol = 6.9634*10**(8)
r_ana = 6.9634*10**(6)
r_neu = 11000
r_bh = 2*m_bh*G/c**2


lua = weak(m_lua, r_lua)
jup = weak(m_jup, r_jup)
sol = weak(m_sol, r_sol)
ana = weak(m_ana, r_ana)
neu = weak(m_neu, r_neu)
bh = weak(m_bh, r_bh)

print(f"Lua: {lua.angDesv():>18.3e}")
print(f"J√∫piter: {jup.angDesv():>14.3e}")
print(f"Sol: {sol.angDesv():>18.3e}")
print(f"An√£-Branca: {ana.angDesv():>11.3e}")
print(f"Neutrons: {neu.angDesv():>13.3e}")
print(f"Buraco Negro: {bh.angDesv():.3e}")
```

```python
plt.figure(figsize=(10,5))
plt.yscale("log")
# plt.xscale("log")
plt.plot(m_lua, lua.angDesv(), 'ro', label="Lua")
plt.plot(m_jup, jup.angDesv(), 'bo' , label="J√∫piter")
plt.plot(m_sol, sol.angDesv(), 'go' , label="Sol")
plt.plot(m_ana, ana.angDesv(), 'co' , label="An√£ Branca")
plt.plot(m_neu, neu.angDesv(), 'mo' , label="Estrela de Neutrons")
plt.plot(m_bh, bh.angDesv(), 'yo' , label="Buraco Negro")
plt.grid()
plt.minorticks_on()
# plt.ylim(1, 3e5)
plt.ylabel("Desvio Angular (rad)", fontsize = 15)
plt.xlabel("Massa (Kg)", fontsize = 15)
plt.legend(loc=5, fontsize=15)
plt.title("Desvio Angular em fun√ß√£o da massa", fontsize=15)

# plt.savefig("Desvio2.pdf")
```
