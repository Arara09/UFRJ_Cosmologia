# Questão 2

Espaços de Lobachevski bidimensionais podem ser visualizados como hiperbolóides em um
espaço 3-D Lorentziano (isto é, com elemento de linha com uma das dimensões com sinal
trocado). Verifique que embutindo a superfı́cie x2 + y 2 − z 2 = −a2 em um espaço com
elemento de linha ds2 = dx2 + dy 2 − dz 2 resulta em um espaço de Lobachevski. Faça também
um gráfico desta superfı́cie. [Ref.: seção 1.3 do livro do Mukhanov.]

```python

a = 5
r = 30

x = np.arange(-r, r + 0.1, 0.1)
y = np.arange(-r, r + 0.1, 0.1)

# x, y = np.meshgrid(x, y)
z1 = np.sqrt(x**2 + y**2 + a**2)
z2 = -np.sqrt(x**2 + y**2 + a**2)
```

```python
fig, ax = plt.subplots(figsize=(10,10), subplot_kw={"projection": "3d"})
ax.view_init(elev=-15, azim=20, roll=0)
ax.plot_surface(x, y, z1, cmap=cm.coolwarm, edgecolor='royalblue',lw=0.5)
ax.plot_surface(x, y, z2, cmap=cm.coolwarm_r, edgecolor='royalblue',lw=0.5)
ax.set_zlim(-20, 50)

ax.plot([], [], [],'' , label=r"$x^2+y^2-z^2 = -a^2$")
ax.set_xlabel("x",fontsize=20)
ax.set_zlabel('Z',fontsize=20)
ax.set_ylabel("y",fontsize=20)
ax.set_title("Hiperboloide", fontsize=30)
ax.legend(fontsize=15,loc='center right')
# plt.savefig("Hiperboloide.pdf")
plt.show()
```

## Letra b)
Discorra sobre como deveria ser a distribuição de energia desses fótons e calcule então a
fração f de fótons com energia maior que a energia de ionização do hidrogênio.

```python
E0 = 13.6


x = np.geomspace(0.2, 5, 300)

a = 0.42*(2.7*E0/x)**2
b = np.exp(-2.7*E0/x)

f = a*b
```

```python
plt.close()

plt.figure(figsize=(10,5))
plt.yscale("log")
plt.plot(x,f)
plt.grid()
plt.minorticks_on()
plt.ylabel(r"f (h$\nu$ > 13.6)", fontsize=15)
plt.xlabel("< E > (eV)", fontsize=15)
plt.title("<E>  << -13.6", fontsize=20)
# help(ax.text)
plt.text(0.5,0.3, r"$f \approx 0.42 \left( \frac{2.7E_0}{<E>}\right)^2e^{ -\frac{2.7E_0}{<E>}}$",bbox={'facecolor': 'white', 'alpha': 1, 'pad': 10}, 
         verticalalignment='bottom', horizontalalignment='right', transform=ax.transAxes, fontsize=25)
# plt.savefig("blackbody_1.pdf")
plt.show()
```
