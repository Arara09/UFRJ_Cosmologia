## 1) Introdução

O lenteamento gravitacional é a ideia de que a luz pode ser desviada de sua trajetória devido a gravidade de um objeto, de forma semelhante ao que acontece em lentes no estudo da óptica. Esse efeito da gravidade já havia sido mencionado no livro Opticks de Isaac Newton, publicado em 1704. Durante o século XVIII, outros físicos como John Mitchell, Henry Cavendish e  Laplace também estudaram como o campo gravitacional de objetos massivos poderiam afetar a luz, que até então era pensada como um partícula com massa. Inclusive, eles chegaram a propor que poderia existir um corpo tão massivo que sua velocidade de escape seria maior que a da luz, calculando o que hoje é conhecido como Raio de Schwarzschild.<a name="c1"></a>[<sup>[1]</sup>](#ct1)

Em 1801, Johann Soldner publicou o desvio angular em uma órbita hiperbólica causado por um corpo com massa M. Para uma partícula com velocidade inicial *v* e parâmetro de impacto *b*, teremos:
$$
                \tan{\frac{\delta}{2}} = \frac{GM}{b v^2}
$$
Para um ângulo de deflexão muito pequeno (que é o caso para a maior parte dos objetos), a tangente do ângulo pode ser aproximada para o valor do próprio ângulo. Dessa forma, o desvio é encontrado como:
$$
        \delta = \frac{2 G M}{b v^2}
$$

# Parte 1 - Microlenteamento

O microlenteamento ocorre quando o lenteamento gravitacional é causado por "pequenos" objetos compactos, como por exemplo planetas, estrelas e inclusive buracos negros. Por meio da distorção no espaço-tempo produzida por esses objetos, o microlenteamento é útil para detectar objetos cuja emissãode radiação é baixa ou nula. Como exemplo, os efeitos de microlenteamento podem ser usados para descobrir exoplanetas, estrelas anã marrons e também buracos negros.

Começaremos estudando a deflexão da luz por esses objetos pontuais.

## 2) Luz Corpuscular

Assumindo que o fóton é uma partícula com massa *m*, iremos agora derivar a equação do desvio em sua trajetória provocado por um corpo com massa M, utilizando mecânica clássica. Essa seção é baseada na parte 1.1 do livro do Meneghetti.<a name="c2"></a>[<sup>[2]</sup>](#ct2)
A força que a partícula irá sentir pela presença da lente é dada pela Lei da Gravidade de Newton:
\begin{equation}
            F = \frac{GMm}{r^2}.
\label{eq:law_gravity}
\end{equation}
Já pela Segunda Lei de Newton, a força exercida sobre a partícula irá gerar uma aceleração nela. Essa relação é dada pela fórmula:
\begin{equation}
            F = \frac{dp}{dt}= ma.
\label{eq:second_law}
\end{equation}
Iremos também tomar como base o princípio de equivalência, portanto a massa gravitacional e inercial da partícula serão a mesma. 

Dado um fóton com momento inicial *p* na direção x passando próximo do Sol, cujo centro está na coordenada (0, a), a distância do fóton para o centro do Sol será:
\begin{equation*}
            r = \sqrt{x^2 + (y-a)^2},
\end{equation*}
dado que *x* e *y* são as coordenadas do fóton em um dado instante. O momento do fóton na análise clássica pode ser escrito como $p = mc$, portanto, ao combinar as equações da gravitação e da Segunda Lei, e substituindo a massa do fóton por $m = p/c$, teremos:

\begin{align}
                F_x &= \frac{dp}{dt}cos(\theta)  = \frac{GMm}{r^2}cos(\theta) = \frac{GMp}{cr^2}\frac{x}{r} = \frac{GMp}{c}\frac{x}{(x^2 + (y-a)^2)^{3/2}}\\
                F_y &= \frac{dp}{dt}sin(\theta) = ... = \frac{GMp}{c}\frac{a - y}{(x^2 + (y-a)^2)^{3/2}}
\label{eq:force_comp}
\end{align}
   


Como a partícula começa na direção *x* e sofrerá um desvio pequeno, podemos considerar que $dx = cdt$. Logo, podemos escrever as derivadas temporais dos momentos como:
\begin{equation*}
            \frac{dp_i}{dt} = \frac{dp_i}{dx}\frac{dx}{dt} = c \frac{dp_i}{dx}.
\end{equation*}
Igualando às componentes de força:
\begin{align*}
            \frac{dp_x}{dx} &= \frac{GMp}{c^2}\frac{x}{(x^2 + (y-a)^2)^{3/2}}\\
            \frac{dp_y}{dx} &= \frac{GMp}{c^2}\frac{a-y}{(x^2 + (y-a)^2)^{3/2}}
\end{align*}
Integrando as equações de menos infinito até o infinito na coordenada *x*, temos a variação que cada componente do momento sofrerá do momento inicial (antes da interação) até o momento final (depois da interação). Para a componente $p_x$:

\begin{equation*}
            \Delta p_x = \frac{GMp}{c^2}\int_{-\infty}^{\infty}\frac{x}{(x^2 + (y-a)^2)^{3/2}}dx
\end{equation*}

Fazendo a substituição $u = (x^2 + (y-a)^2)^{1/2} $, $du = xdx/u$

\begin{equation*}
            \Delta p_x = \frac{GMp}{c^2}\int_{-\infty}^{\infty}\frac{udu}{u^3} = \frac{GMp}{c^2}\int_{-\infty}^{\infty}\frac{du}{u^2} = \frac{GMp}{c^2}\left(\frac{-1}{u}\right)\bigg|_{-\infty}^{\infty} = 0
\end{equation*}

Agora para a componente $p_y$, fazendo a substituição $x = (y-a)\tan(\theta)$, $dx = (y-a)\sec^2(\theta)d\theta$:

\begin{equation*}
            \Delta p_y = \frac{GMp}{c^2}\int_{-\infty}^{\infty}\frac{a-y}{(x^2 + (y-a)^2)^{3/2}}dx = \frac{GMp}{c^2}\int_{-\pi/2}^{\pi/2}\frac{-(y-a)}{(y-a)^3(1 + \tan^2(\theta))^{3/2}}(y-a)\sec^2(\theta)d\theta 
\end{equation*}

\begin{equation*}
             \Delta p_y=  \frac{-GMp}{c^2}\frac{1}{(y-a)}\int_{-\pi/2}^{\pi/2}\frac{\sec^2(\theta)d\theta}{\sec^3(\theta)} = \frac{-GMp}{c^2}\frac{1}{(y-a)}\int_{-\pi/2}^{\pi/2}\cos(\theta)d\theta =  \frac{GMp}{c^2}\frac{1}{(a-y)} \sin(\theta)\bigg|_{-\pi/2}^{\pi/2}
\end{equation*}

\begin{equation}
            \Delta p_y = \frac{2GMp}{c^2}\frac{1}{(a-y)}
\end{equation}

Com base na figura acima, podemos calcular o ângulo de desvio como:
\begin{equation}
          \sin(\Psi) \approx  \Psi = \frac{\Delta p_y}{p} = \frac{2GM}{c^2}\frac{1}{a-y}
\end{equation}
Dessa forma, dado uma lente com massa M eraio R e dado que o fóton tangencia a lente, o ângulo de deflexão no caso clássico para um fóton corpuscular é:

\begin{equation}
            \Psi = \frac{2GM}{Rc^2}
\label{eq:corp} \tag{7}
\end{equation}

### 2.1) Cálculo dos desvios

Para começar a parte computacional, iremos primeiro importar algumas bibliotecas úteis para o cálculo matemático e também para a produção de gráficos. Aproveitaremos também para definir algumas variáveis globais que serão utilizadas frequentemente.

```python
import numpy as np
import math
from scipy import special as sy
from matplotlib import pyplot as plt
%matplotlib inline                      
# Projetar os gráficos no próprio notebook


c = 299792458                 # Velocidade da luz
G = 6.6743*10**(-11)          # Constante Gravitacional
```

Será interessante definir algumas variáveis, para a massa e o raio, de objetos que utilizaremos constantemente. Inicialmente iremos definir variáveis para a Lua, Júpiter, Sol, uma Anã Branca típica, uma Estrela de Nêutros e um Buraco Negro com 50 massas solares.

```python
m_moon   , r_moon    =    7.35e22   ,  1.7374e6
m_jup    , r_jup     =   1.898e27   ,  6.9911e7
m_sun    , r_sun     =   1.989e30   ,  6.9634e8
m_WD     , r_WD      = 0.6 * m_sun  ,  6.9634e6                 # White Dwarf
m_NS     , r_NS      = 1.5 * m_sun  ,  11000                    # Neutron Star
m_BH     , r_BH      = 50. * m_sun  ,  100. * m_sun * G / c**2  # Black Hole
```

Agora será definida a classe para realizarmos o cálculo do desvio no caso da luz corpuscular. Para tanto, iremos registrar as informações da massa dentro da classe, e depois utilizaremos ela junto com o parâmetro de impactos para construir o método angDef (*Angular Deflection*). Calcularemos o maior desvio ângular possível, para quando o fóton tangencia a superfície do objeto, portanto o parâmetro de impacto será o próprio raio do objeto. O desvio angular é dado pela equação encontrada acima e o resultado poderá ser exibido em radianos ou em arcosegundos.

```python
class Corpuscle():
    
    def __init__(self, M):
        self.M = M
    
    def angDef(self, r, arcsec = False):
        
        angle = 2 * G * self.M / (c**2 * r)
        
        if arcsec:
            return angle * 180 * 60 * 60 / np.pi
            
        else:
            return angle
        
    def show(self, r, name):
        print(f"{str(name):<20} - Ângulo de Deflexão: {self.angDef(r):.3e} radianos ({self.angDef(r, arcsec=True):.3f}\")")
        
```

Agora com a classe pronta, faremos o cálculo para cada um dos objetos astronômicos definidos acima.

```python
moon_Corp = Corpuscle(m_moon)
moon_Corp.show(r_moon, "Lua")

jup_Corp = Corpuscle(m_jup)
jup_Corp.show(r_jup, "Júpiter")

sun_Corp = Corpuscle(m_sun)
sun_Corp.show(r_sun, "Sol")

wd_Corp = Corpuscle(m_WD)
wd_Corp.show(r_WD, "Anã Branca")

ns_Corp = Corpuscle(m_NS)
ns_Corp.show(r_NS, "Estrela de Nêutrons")

bh_Corp = Corpuscle(m_BH)
bh_Corp.show(r_BH, "Buraco Negro")
```

Os valores calculados mostram que o ângulo de deflexão é muito pequeno para a maior parte dos objetos. Importante notar também que a Anã Branca, mesmo tendo uma massa menor que a do Sol, produziu um desvio muito maior por ser um objeto mais denso, o que permite o parâmetro de impacto ser bem menor. Já a Estrelas de Nêutrons e o Buraco Negro produziram desvio de ordem muito maior que os outros, mesmo no caso corpuscular.

Logo, iniciaremos a conta pela métrica de Schwarzschild:

### 3.1) Regime Fraco

Consideraremos primeiro a situação de campo fraco. Nesse caso, estamos considerando que:

\begin{equation*}
        \Phi/c^2 = -\frac{GM}{rc^2} << 1 
\end{equation*}

Ou seja, a massa M não causará uma grande distorção na métrica e, portanto, podemos considerar somente uma pertubação de primeira ordem. Utilizando a expansão binomial nos termos da métrica de Schwarzschild, teremos que:

\begin{equation*}
                \left( \frac{1 + \frac{\Phi}{2c^2} }{1 - \frac{\Phi}{2c^2}} \right)^2 = \left(1 + (2)\left(\frac{\Phi}{2c^2} \right) + ... \right) \left( 1 + (-2)\left(\frac{-\Phi}{2c^2}\right) + ... \right) \approx 1 + \frac{2\Phi}{c^2}
\end{equation*}

\begin{equation*}
            \left(1 - \frac{\Phi}{2c^2}  \right)^{4} \approx 1 - 4 \left(\frac{\Phi}{2c^2} \right) = 1 - \frac{2\Phi}{c^2} 
\end{equation*}

Substituindo essas aproximações na métrica:

\begin{equation}
          ds^2 = \left( 1 + \frac{2\Phi}{c^2} \right)c^2 dt^2 - \left(1 - \frac{2\Phi}{c^2} \right)(dx^2 + dy^2 + dz^2)
\end{equation}

A luz percorre o que é conhecido como geodésica nula, o que significa que $ds^2 = 0$. Igualando a equação acima a zero e reorganizando os termos, podemos calcular a velocidade do fóton de luz nesse campo fraco:

\begin{equation*}
            \left( 1 + \frac{2\Phi}{c^2} \right)c^2 dt^2 = \left(1 - \frac{2\Phi}{c^2} \right)(dx^2 + dy^2 + dz^2)
\end{equation*}

\begin{equation*}
        c' = \frac{dx'}{dt} = c \sqrt{\frac{1 + \frac{2\Phi}{c^2} }{1 - \frac{2\Phi}{c^2} }} = c \left( 1 + (1/2)\frac{2\Phi}{c^2} + ... \right) \left(1 - (-1/2)\frac{2\Phi}{c^2} + ... \right) \approx c \left(1 + \frac{2\Phi}{c^2} \right)
\end{equation*}

Com o valor da velocidade da luz no campo gravitacional, podemos encontrar um índice de refração efetivo pela razão entre as velocidades.

\begin{equation*}
        n = \frac{c}{c'} = \frac{1}{1 + \frac{2\Phi}{c^2}} \approx 1 - \frac{2\Phi}{c^2}
\end{equation*}

Para obter o ângulo de deflexão provocado, vamos verificar o caminho percorrido pelo fóton por meio do princípio de Fermat. O príncipio dita que o caminho que a luz percorre  entre dois pontos é aquele que requer o menor tempo. Matematicamente, isso pode ser traduzido como:

\begin{equation*}
        \delta \int_{x_0}^{x_1}n[\vec{x}(l)] dl = 0
\end{equation*}

Parametrizando a curva percorrida pelo fóton ($x(\lambda)$), podemos trocar a variável de integração:

\begin{equation*}
        \delta \int_{\lambda_0}^{\lambda_1}n[\vec{x}(\lambda)] \bigg| \frac{d\vec{x}}{d\lambda} \bigg| d\lambda
\end{equation*}

Chamando $\vec{x}' = \frac{d\vec{x}}{d\lambda}$ e $L = n[\vec{x}(\lambda)] \left| \frac{d\vec{x}}{d\lambda} \right|$, a equação de Euler para esse problema ficará:

\begin{equation*}
        \frac{d}{d\lambda}\left(\frac{\partial L}{\partial \vec{x}'}\right) - \frac{\partial L}{\partial \vec{x}} = 0
\end{equation*}

\begin{equation*}
        \frac{d}{d\lambda}\left(n \frac{\vec{x}'}{|\vec{x}'|}\right) - |\vec{x}'|\frac{\partial n}{\partial \vec{x}} = \frac{d }{d\lambda}\left(n \frac{\vec{x}'}{|\vec{x}'|}\right) - |\vec{x}'|\vec{\nabla} n =  0
\end{equation*}

Como o parâmetro $\lambda$ até então é um parâmetro arbitrário, podemos utilizar um que normaliza o vetor tangente ($|\vec{x}'| = 1$). Com essa simplificação, teremos que:

\begin{equation*}
        \frac{d }{d\lambda}\left(n \hat{x}'\right)  = \left(\frac{dn}{d\vec{x}}\cdot \frac{d\vec{x}}{d\lambda}\right) \vec{x}' + n \vec{x}'' =(\vec{\nabla}n \cdot \hat{x}')\hat{x}' + n \vec{x}''  = \vec{\nabla} n
\end{equation*}

\begin{equation*}
        n \vec{x}'' = \vec{\nabla} n - (\vec{\nabla}n \cdot \hat{x}')\hat{x}'
\end{equation*}

O lado direito da integral é o gradiente do índice de refração menos a componente paralela ao versor da derivada. Portanto, podemos chamar esse termo do gradiente perpendicular do índice de refração. Passando o 'n' dividindo:

\begin{equation*}
        \vec{x}'' =\frac{1}{n} \vec{\nabla}_\perp n = \vec{\nabla}_\perp \ln n
\end{equation*}

Retomando a definição de 'n' acima,

\begin{equation*}
        \vec{x}'' = \vec{\nabla}_\perp ln \left( 1 -\frac{2\Phi}{c^2} \right) = \vec{\nabla}_\perp\left[ -\frac{2\Phi}{c^2} + ...\right] \approx -\frac{2}{c^2}\vec{\nabla}_\perp \Phi
\end{equation*}

Integrando as equações, obtemos uma fórmula para o ângulo de deflexão:

\begin{equation*}
            \vec{\alpha} = \vec{x}'_{in} - \vec{x}'_{out} = \frac{2}{c^2} \int_{\lambda_0}^{\lambda_1} \vec{\nabla}_\perp \Phi d\lambda
\end{equation*}

#### Aproximação de Born

Pela equação acima, seria necessário conhecer o caminho percorrido pela partícula para poder obter o ângulo de deflexão. Contudo, para o caso do regime fraco ($\Phi /c^2 << 1$), podemos considerar que o ângulo de deflexão é pequeno o suficiente para realizar a integração no caminho que o fóton percorreria na ausência da lente. Essa aproximação é conhecida na teoria de espalhamento como Aproximação de Born.

Portanto, definindo a direção que conecta a fonte com o observador como $\hat{z}$, realizaremos a integração na variável z:

\begin{equation*}
           \vec{\alpha} = \frac{2}{c^2} \int_{-\infty}^{\infty} \vec{\nabla}_\perp \Phi dz = \frac{2}{c^2} \int_{-\infty}^{\infty}\left( \frac{\partial}{\partial r} \hat{r} + \frac{1}{r}\frac{\partial}{\partial \phi} \hat{\phi} \right) \Phi dz
\end{equation*}

O gradiente perpendicular foi tomado então nas coordenadas polares. Como o potencial gravitacional só depende da distância, teremos então:


\begin{equation*}
            \vec{\alpha} = \frac{2}{c^2} \hat{r}  \int_{-\infty}^{\infty}\frac{\partial \Phi}{\partial r}  dz=  \frac{-2GM}{c^2}  \hat{r} \int_{-\infty}^{\infty} \frac{\partial}{\partial r}\left( \frac{1}{\sqrt{r^2 + z^2}} \right)  dz  =  \frac{2GM}{c^2}  \hat{r} \int_{-\infty}^{\infty} \frac{r}{(r^2 + z^2)^{3/2}}   dz 
\end{equation*}

\begin{equation*}
            \vec{\alpha} = \frac{2GMr}{c^2}  \hat{r} \left( \frac{z}{\sqrt{r^2 + z^2}} \right)   \bigg|_{-\infty}^{\infty}  =   \frac{2GM r}{c^2} \frac{2}{r^2} \hat{r}
\end{equation*}

Portanto, encontramos que o ângulo de deflexão para uma massa pontual, no regime fraco, é dada por:

\begin{equation*}
            \vec{\alpha} = \frac{4GM}{c^2 r} \hat{r}
\end{equation*}

Comparando com a equação encontrada na seção anterior, percebemos que ao considerar a Relatividade Geral o desvio maior por um fator de 2. Essa diferença considerável no valor obtido pelas duas teorias foi o que permitiu verificar experimentalmente o resultado da Relatividade Geral no eclipse de 1919.

#### Cálculo dos Desvios

Agora, iremos repetir o procedimento anterior para o caso relativístico. Criaremos uma classe para definir novos objetos e realizar o cálculo com a nova fórmula.

```python
class WeakField:
    
    def __init__(self, M):
        self.M = M
        
    def angDef(self, r, arcsec = False):
        
        angle = 4 * G * self.M / (c**2 * r)
        
        if arcsec:
            return angle * 180 * 60 * 60 / np.pi
            
        else:
            return angle
        
    def show(self, r, name):
        print(f"{str(name):<20} - Ângulo de Deflexão: {self.angDef(r):.3e} radianos ({self.angDef(r, arcsec=True):.3f}\")")
    
```

Vamos reutilizar os seis objetos astronômicos usados no exemplo anterior e mostrar os novos ângulos de deflexão ao tangenciar o objeto.

```python
moon_WF = WeakField(m_moon)
moon_WF.show(r_moon, "Lua")

jup_WF = WeakField(m_jup)
jup_WF.show(r_jup, "Júpiter")

sun_WF = WeakField(m_sun)
sun_WF.show(r_sun, "Sol")

wd_WF = WeakField(m_WD)
wd_WF.show(r_WD, "Anã Branca")

ns_WF = WeakField(m_NS)
ns_WF.show(r_NS, "Estrela de Neutrons")

bh_WF = WeakField(m_BH)
bh_WF.show(r_BH, "Buraco Negro")
```

O ângulo de deflexão é o dobro do calculado com a classe anterior, como queríamos. Podemos perceber que no caso da Estrela de Neutrons e do Buraco Negro, o valor do desvio já está na ordem de radianos. Como essas são algumas das classes de objetos mais densas do universo, o fóton que os tangencia já não está mais no regime de gravitação fraca. Portanto, será necessário realizar o cálculo completo para o regime forte.

### 3.2) Regime Forte

A grande maioria dos objetos astronômicos podem ser tratados no regime fraco, pois eles não chegam a ser tão compactos a ponto de necessitar uma correção. Como um exemplo, um aglomerado de galáxias, mesmo podendo ultrapassar uma massa de $10^{15}M_\odot$, tem um potencial de aproximadamente $|\Phi| \approx 10^{-4}c^2$ <a name="c3"></a>[<sup>[3]</sup>](#ct3). Contudo, casos como o de uma Estrela de Neutrons e Buracos Negros, dependendo da distância, podem produzir campos gravitacionais muito fortes e grandes deflexões na trajetória da luz.

As contas para esse regime são mais elaboradas, portanto não apresentaremos o seu desenvolvimento e seguiremos diretamante para a solução do ângulo de deflexão e a implementação em código. Primeiramente, devemos tratar da métrica no caso genérico com simetria esférica e estacionária:

\begin{equation*}
               ds^2 = A(r)c^2dt^2 - B(r)dr^2 - C(r)(d\theta^2 + \sin^2\theta d\phi^2)
\end{equation*}

Para $A(r) = 1 - 2GM/rc^2 $, $B(r) = A(r)^{-1}$ e $C(r) = r^2$, temos a métrica de Schwarzschild em coordenadas esféricas. Essas funções serão importante para transformar do valor $r_m$, que é a menor distância entre o fóton e a lente, para o parâmetro de impacto $u$:

\begin{equation*}
                u^2 = \frac{C(r_m)}{A(r_m)}
\end{equation*}

A equação exata para o ângulo de deflexão com a métrica de Schwarzschild foi encontrada em 1959 por Charles Darwin (neto do famoso biólogo).

\begin{equation*}
                \alpha  = - \pi + 4 \sqrt{ \frac{r_m}{s} } F(\varphi, m)
\end{equation*}

Onde $F(\phi, m)$ é a integral elíptica de primeira espécie, e:

\begin{align*}
          s &= \sqrt{\left(r_m - \frac{2GM}{c^2} \right) \left (r_m + \frac{6GM}{c^2} \right ) }\\
          m &= \frac{s - r_m + \frac{6GM}{c^2}}{2s}\\
          \varphi &= \arcsin{\sqrt{\frac{2s}{3r_m - \frac{6GM}{c^2} + s}}}
\end{align*}

### Análise da equação de desvio angular

Antes de calcular os devios corrigidos, iremos analisar os limites da equação de desvio angular apresentada.

Pensando, primeiro, em um fóton que passará muito distante da lente gravitacional, teremos que $r_m \approx u \longrightarrow \infty$. Nesse caso, as variáveis $s$, $m$ e $\varphi$ ficarão:

\begin{align*}
          s &\approx r_m\\
          m &\approx \frac{s - r_m}{2s} \approx 0\\
          \varphi &\approx \arcsin{\sqrt{\frac{2s}{3r_m + s}}} \approx \arcsin{\frac{1}{\sqrt{2}}} = \frac{\pi}{4}
\end{align*}
Usando a definição da integral elíptica de primeira espécie, temos que:
\begin{equation*}
           F(m,\varphi) = \int_0^\varphi \frac{d\theta}{\sqrt{1-m\sin^2\theta}}
\end{equation*}
\begin{equation*}
           F\left(0,\frac{\pi}{4}\right) = \int_0^{\frac{\pi}{4}} d\theta = \frac{\pi}{4}
\end{equation*}

Substituindo as variáveis no cálculo do desvio, encontramos:

\begin{equation*}
               \alpha = -\pi + 4 (1)\left(\frac{\pi}{4}\right) = 0
\end{equation*}
Como esperado, quando o fóton passa a uma grande distância da lente, não há deflexão e o fóton segue na mesma direção. Porém, com esse teste já podemos confirmar um erro na equação apresentada no livro do Meneghetti (equação 42 do capítulo "Light Deflection"). Na equação do livro, aparece $G/c^2$ multiplicando o termo com a integral elíptica, só que, além do problema dimensional que isso geraria (pois o ângulo $\alpha$ é adimensional), também estaríamos diminuindo o segundo termo por uma ordem de $10^{-27}$. Por isso, a equação do ângulo de deflexão já foi apresentada acima sem este termo. Além dela, também foram realizados ajustes nas outras equações para a variável $s$ tivesse unidade de comprimento e que $m$ e $\varphi$ fossem adimensionais.

Outro limite que pode ser calculado é quando o fóton se aproxima muito da lente. Se colocássemos $r_m \longrightarrow r_s$, ou seja, se o fóton se aproximasse a ponto de tangenciar o raio de Schwarzschild ($r_s = 2GM/c^2$), teríamos $s \longrightarrow 0$ e $m \longrightarrow \infty$, o que causaria uma indeterminação. Isso ocorre pois o fóton irá espiralar para dentro do buraco negro, sem conseguir escapar. Na realidade, para todos valores de $r_m < 1,5 r_s$ o fóton será eventualmente capturado pelo buraco negro.

Para a distância $r_m = 1,5 r_s$, temos o que chamado de esfera de fótons, uma região de órbitas instáveis de fótons, onde teoricamente o fóton poderia circular o buraco negro indefinidamente. Caso a aproximação máxima do fóton seja um pouco acima dessa distância, o fóton será desviado, mas escapará do buraco negro. Para aproximações do fóton abaixo do valor crítico, ele irá espiralar até cair no buraco negro. Essa distância da esfera de fótons é equivalente a um parâmetro de impacto $u = \frac{3\sqrt{3}}{2}r_s$.


### Cálculo dos Desvios

Seguindo o mesmo procedimento, iremos criar uma classe para calcular todas as funções acima e utilizá-las para determinar o desvio angular no caso de campo forte.

```python
class StrongField:
    
    def __init__(self, M):
        self.M = M
        self.r_s = 2 * G * self.M / c**2              # Raio de Schwarzschild
        
    
    def A(self, r):
        return 1 - (self.r_s / r )
    
    
    def B(self, r):
        return self.A(r)**(-1)
    
    
    def C(self, r):
        return r**2
    
    
    def u(self, r):                                   # Parâmetro de Impacto
        return np.sqrt(self.C(r)/self.A(r))
    
    
    def s(self, r):
        return np.sqrt( (r - self.r_s) * (r + 3*self.r_s) )
    
    
    def m(self, r):
        return (self.s(r) - r + 3*self.r_s)/(2 * self.s(r))
    
    
    def phi(self, r):
        f1 =  (2*self.s(r)) / (3*r - 3*self.r_s + self.s(r))
        f2 = np.sqrt(f1)
        return np.arcsin(f2)
    
    
    def angDef(self, r, arcsec = False):
        s   = self.s(r)
        m   = self.m(r)
        phi = self.phi(r)
        
        
        F = sy.ellipkinc(phi, m)
        
        angle = -np.pi + 4* np.sqrt( r / s ) * F
        
        if arcsec:
            return angle * 180 * 60 * 60 / np.pi
            
        else:
            return angle
        
    
    def show(self, r, name):
        print(f"{str(name):<20} - Ângulo de Deflexão: {self.angDef(r):.3e} radianos ({self.angDef(r, arcsec=True):.3f}\")")
```

```python
moon_SF = StrongField(m_moon)
moon_SF.show(r_moon, "Lua")

jup_SF = StrongField(m_jup)
jup_SF.show(r_jup, "Júpiter")

sun_SF = StrongField(m_sun)
sun_SF.show(r_sun, "Sol")

wd_SF = StrongField(m_WD)
wd_SF.show(r_WD, "Anã Branca")

ns_SF = StrongField(m_NS)
ns_SF.show(r_NS, "Estrela de Neutrons")

bh_SF = StrongField(m_BH)
bh_SF.show(1.6 * r_BH, "Buraco Negro")
```

 Comparando com os valores calculados no regime fraco, percebemos que para a Lua, Júpiter e pro Sol o valor continuou o mesmo, o que significa que a correção é muito pequena para esses objetos. Para a Anã Branca, teve uma mudança na terceira casa decimal, enquanto que a deflexão da Estrela de Nêutrons, que já é um objeto muito denso, quase dobrou o ângulo de deflexão ao tangenciar. ## 4) Efeitos do Microlenteamento

Uma equação importante para o estudo do lenteamento gravitacional é a equação da lente, que relaciona os ângulos e as distâncias envolvidas. Estamos considerando que todos os ângulos envolvidos são pequenos, para que a relação seja válida.

\begin{equation*}
            \vec{\beta} D_S = \vec{\theta} D_S - \vec{\alpha} D_{LS}(\vec{\theta})
\end{equation*}



Na equação, a distância $D_S$ é a distância diâmetro ângular entre o observador e a fonte, enquanto $D_{LS}$ é a distância diâmetro angular entre a lente e a fonte, pois o ângulo de deflexão é medido no plano da lente. O ângulo $\vec{\beta}$ é o ângulo entre o eixo e a fonte, $\vec{\theta}$ é o ângulo entre o eixo e a imagem (a direção que o observador vê a luz) e $\vec{\alpha}$ é o ângulo entre a fonte e a imagem no plano da lente. O que a figura chama de $\vec{\hat\alpha}$ é o ângulo de deflexão reduzido, ou seja, o ângulo entre a posição real da fonte e a imagem da posição do observador. Esses dois ângulos podem ser relacionados por:

\begin{equation*}
        \vec{\hat\alpha} = \frac{D_{LS}}{D_S} \vec{ \alpha}
\end{equation*}
 
Dessa forma, a equação da lente reduz para:
 
\begin{equation*}
            \vec{\beta} = \vec{\theta} - \vec{\hat\alpha}(\vec{\theta})
\end{equation*}

Como vimos acima, o ângulo de deflexão para massa pontuais é dado por: 

\begin{equation*}
            \vec{ \alpha} = \frac{4GM}{c^2 \xi}\hat r
\end{equation*}

Onde chamamos o parâmetro de impacto de $\xi$ para igualar com a figura. Como os ângulo envolvidos são pequenos e lembrando da relação $\vec{\xi} = D_L \vec{\theta}$ (e para o plano da fonte $\vec{\eta} = D_S \vec{\beta}$), temos:

\begin{equation*}
             \vec{\beta} = \vec{\theta} - \frac{D_{LS}}{D_S D_L}\frac{4GM}{c^2 \theta^2}\vec{\theta}
\end{equation*}

Podemos também escrever a equação escalar:

\begin{equation*}
                \beta = \theta - \frac{D_{LS}}{D_S D_L}\frac{4GM}{c^2 \theta}
\end{equation*}

Podemos perceber, portanto, que essa equação não é linear e ocorrerá também a aparição de múltiplas imagens. Um caso especial é o de uma fonte alinhada com o centro da lente e o observador($\beta = 0$), nessa situação, resolvendo para $\theta$ teremos:

\begin{equation*}
                \theta_E = \sqrt{ \frac{4GM}{c^2}\frac{D_{LS}}{D_S D_L}}
\end{equation*}

Essa quantidade é chamada de Raio de Einstein, pois nesse caso a imagem da fonte se distorceria em um círculo ao redor da lente. Retornando à equação da lente, vamos definir novas coordenadas em unidades do Raio de Einstein: $x = \theta/\theta_E$ e $y = \beta/\theta_E$. Solucionando a equação para um valor de $\beta$ qualquer diferente de zero, encontramos:

\begin{equation*}
                x_{\pm} = \frac{1}{2} \left (y \pm \sqrt{y^2 + 4} \right)
\end{equation*}

Como podemos ver, a imagem $x_-$ sempre estará dentro do círculo de raio 1 (ou Anel de Einstein), enquanto que a imagem $x_+$ sempre estará do lado de fora, diametralmente oposta à primeira imagem. A seguir reproduziremos um gráfico do Meneghetti para o caso de várias fontes alinhadas no eixo horizontal.

```python
def images(y):
    "Calcula a posição das imagens no microlenteamento"
    
    x1 = (1/2)*(y - np.sqrt(y**2 + 4))
    x2 = (1/2)*(y + np.sqrt(y**2 + 4))
    
    return x1, x2

phi = np.linspace(0, 2*np.pi, 200)
circ_x = np.cos(phi)                        # Monta o círculo de raio 1
circ_y = np.sin(phi)


# Configurações do plot
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].set_xlim(-2.5, 2.5)
axes[0].set_ylim(-2.5, 2.5)
axes[1].set_xlim(-2.5, 2.5)
axes[1].set_ylim(-2.5, 2.5)

axes[0].set_title("Plano das Fontes")
axes[0].set_xlabel("y1")
axes[0].set_ylabel("y2")

axes[1].set_title("Plano da Lente")
axes[1].set_xlabel("x1")
axes[1].set_ylabel("x2")


axes[0].plot(circ_x, circ_y)
axes[0].plot(0, 0, '*r', markersize = 15)            # Plota a Lente e o Círculo nos dois planos
axes[1].plot(circ_x, circ_y)
axes[1].plot(0, 0, '*r', markersize = 15)


for i in range(1, 20, 2):
    im1, im2 = images(i/10.)
    
    axes[0].plot(i/10., 0, 'o')
    axes[1].plot([im1, im2], np.zeros(2), 'o')       # Plota as fontes e as imagens
    
plt.show()
```

Como podemos observar, para fontes muito próximas da lente, as imagens produzidas serão praticamente simétricas e distantes da posição real da fonte. Para fontes distantes da lente, a imagem $x_-$ se aproxima cada vez mais da origem, enquanto que a imagem $x_+$ se aproxima cada vez mais da posição real da fonte.

# Parte 2 - Modelos de Lente Extensa

A seguir trataremos das lentes extensas, que como o nome diz, é um modelo que considera distribuições de massa extensa. Os modelos de lentes extensas são boas aproximações para o lenteamento causado por galáxias e aglomerados de galáxias.

## 5) Esfera Isotérmica Singular

Um modelo simples de lente axialmente simétrica, que é muito utilizado, é a Esfera Isotérmica Singular. Nesse modelo, assumimos que todas as estrelas e outros componentes de matéria que compõem a lente se comportam como partículas de um gás ideal. Assumimos também que essa esfera estará em equilíbrio térmico e hidrostático. Uma caso simples para essas condições é:

\begin{equation*}
            \rho(r) = \frac{\sigma_v^2}{2\pi G r^2} 
\end{equation*}

Onde $\sigma_v$ é a dispersão de velocidade das "partículas". Projetando a densidade no plano da lente, temos uma densidade superficial:

\begin{equation*}
            \Sigma(\xi) = \frac{\sigma_v^2}{2G \xi}
\end{equation*}

### Definições

Antes de prosseguir, devemos definir algumas quantidade que são importantes para o estudo do lenteamento e para o caso de lentes extensas em específico. Primeiramente, é necessário apresentar o potencial efetivo da lente, uma quantidade importante que caracteriza as distribuições extensas de matéria. Ele é definido como a projeção do potencial Newtoniano no plano da lente,  multiplicado por certas constantes:

\begin{equation*}
            \hat\psi (\vec\theta) = \frac{D_{LS}}{D_S D_L}\frac{2}{c^2}\int_{-\infty}^{\infty}\Phi(D_L \vec \theta, z) dz
\end{equation*}

O potencial é definido dessa forma para que uma de suas propriedades seja:

\begin{equation*}
            \vec\nabla  \hat\psi (\vec\theta) = \hat \alpha
\end{equation*}

Ou seja, o gradiente do potencial efetivo é o ângulo de reflexão reduzido, que é medido a partir do observador. Outra propriedade importante do potencial efetivo é o seu laplaciano:

\begin{equation*}
            \nabla^2 \hat\psi = 4\pi G \frac{D_{LS} D_L}{D_S} \frac{2}{c^2}\Sigma (\theta) = 2\frac{\Sigma (\theta)}{\Sigma_{crit}} = 2\kappa
\end{equation*}

Onde $\Sigma_{crit}$ é a densidade superficial crítica, que serve como linha divisória entre lentes "fracas" ($\Sigma < \Sigma_{crit}$) e lentes "fortes"($\Sigma > \Sigma_{crit}$). A quantidade $\kappa$ é a densidade superficial adimensional, também muito conhecida como convergência. Esse nome vem porque a imagem sofrerá uma magnificação isotrópica com relação à fonte, ou seja, ela terá a mesma forma que a fonte, porém aumentada.

Também é importante mencionar a o cisalhamento e a magnificação. Para tanto, vamos começar apresentando a matriz jacobiana A que fará o mapeamento dos pontos no plano da fonte para o plano da lente e descreverá, portanto, a distorção causada nas imagens.
\begin{equation*}
         A  \equiv \frac{\partial \vec\beta}{\partial \vec \theta} = \left( \delta_{ij} - \frac{\partial^2 \hat \psi}{\partial \theta_i \partial \theta_j}\right)  
\end{equation*}

Definindo a quantidade cisalhamento $\gamma$, com componentes:

\begin{align*}
                \gamma_1 &= \frac{1}{2} \left(\psi_{11} - \psi_{22} \right) \\
                \gamma_2 &= \psi_{12} = \psi_{21}
\end{align*}

Podemos, então, reescrever a matriz jacobiana como:


\begin{equation*}
            A = 
            \begin{bmatrix}
                1 - \kappa - \gamma_1 & -\gamma_2\\
                -\gamma_2 & 1 - \kappa + \gamma_1
            \end{bmatrix}
\end{equation*}

O nome cisalhamento é devido à anistropia na imagem causada por essa componente. Dado que $\gamma = \sqrt{\gamma_1^2 + \gamma_2^2}$, os autovalores dessa matriz são: $\lambda_t = 1 - \kappa - \gamma$ e $\lambda_r = 1 - \kappa + \gamma$. Os índices dos autovalores se referem à parte tangencial e radial, que está relacionado com a magnificação das imagens e também com as curvas críticas, que serão apresentada em breve. Começando pela magnificação, ela é definida como o inverso do determinante da matriz jacobiana, pois ela está ligada ao mapeamento dos ângulos sólidos.

\begin{equation*}
            \mu = \frac{1}{det A} = \frac{1}{(1 - \kappa)^2 - \gamma^2}
\end{equation*}

Portanto, para lentes com simetria axial, o inverso dos autovalores serão, respectivamente, a magnificação tangencial e radial. Importante notar a magnificação seria infinita no caso dos autovalores serem zero. Essa condições definem a curva crítica radial e a curva crítica tangencial, que estão presentes no plano da lente. Quando essas curvas são mapeadas para o plano das fontes, elas são chamadas de caústicas. Com exemplo, a lente pontual só possui uma curva crítica (que no caso é a tangencial) que é um círculo de raio $x = 1$ ($\theta = \theta_E$). Essa linha crítica é mapeada em um único ponto no plano das fontes, que é a origem $y = 0$. Portanto, caso uma fonte esteja posicionada em cima da curva cáustica (na origem), ela será deformada sobre a curva crítica (círculo), gerando assim uma imagem no formato de anel, da mesma forma que explicado anteriormente.

Por fim, também há a massa adimensional da lente, dada por:

\begin{equation*}
                m(x) = \frac{M(\xi_0 x)}{\pi \xi_0^2 \Sigma_{crit}}
\end{equation*}

Onde $M(\xi_0 x)$ é a massa contida dentro de um círculo de raio $\xi_o x$. Essa quantidade é útil, pois ela também é utilizada para definir o ângulo de reflexão de uma dada lente.

\begin{equation*}
            \hat\alpha = \frac{m(x)}{x}
\end{equation*}

### Valores para a SIS

Voltando para o caso da Esfera Isotérmica Singular, devemos agora desenvolver as principais propriedades da SIS por meio das quantidade definidas acima. Relembrando a densidade superficial calculada anteriormente para a SIS, por meio de uma escolha de escala angular e de comprimento, podemos reescrever a densidade como:

\begin{equation*}
                    \Sigma(x)  = \frac{\Sigma_{crit}}{2x}
\end{equation*}

Portanto, a convergência será:

\begin{equation*}
            \kappa = \frac{1}{2x}
\end{equation*}

Com a convergência, podemos também determinar a massa adimensional e o ângulo de deflexão.

\begin{equation*}
            m(x) = x \hspace{3cm} \alpha =\frac{x}{|x|}
\end{equation*}

Esse resultado é muito interessante, pois mostra que no caso de uma lente do tipo SIS, a deflexão angular independe do parâmetro de impacto, será sempre 1 (i.e. $\alpha = \theta_E$). Caso a fonte esteja localizada dentro do Raio de Einstein, haverá duas imagens: $x_-  = y -1$ e $x_+ y + 1$. Interessante perceber também que a separação angular entre as duas imagens será sempre $\Delta \theta = 2\theta_E$. Caso a fonte esteja fora do raio, só haverá uma imagem ($x_+ = y + 1$). Essa mudança na quantidade de imagens normalmente ocorre quando a fonte atravessa a cáustica radial, porém o círculo de raio 1 não é propriamente uma cáustica, ele é uma pseudo-cáustica, também conhecida como corte (*cut*). 

A magnificação será:

\begin{equation*}
            \mu(x) = \frac{|x|}{|x| - 1}
\end{equation*}

Portanto, podemos calcular a magnificação de cada imagem, que será:

\begin{equation*}
            \mu(x_+) = 1 + \frac{1}{y} \hspace{3cm} \mu(x_-) = 1 - \frac{1}{y}
\end{equation*}

Portanto, conforme a fonte se aproxima do Raio de Einstein, a imagem $x_-$ se aproxima da origem e diminui até desaparecer.

O cisalhamento na SIS é:

\begin{equation*}
        \gamma = \frac{1}{2x}
\end{equation*}

Com ele podemos calcular as curvas críticas e cáusticas para a SIS. Substituindo na fórmula para os autovalores da matriz jacobiana, temos: $\lambda_t = 1 -x$ e $\lambda_r = 1$. Portanto, não haverá curva crítica radial (pois $\lambda_r \neq 0 $ para qualquer x) e a curva crítica tangencial será um círculo de raio 1. Por $\lambda_r$ ser sempre 1, isso significa que também não teremos magnificação radial (que é o inverso do autovalor), somente tangencial. Mapeando o círculo $x  = 1$ no plano das fontes por meio da equação das lentes, temos que todos os pontos levarão à origem $y = 0$. Portanto, a cáustica tangencial será a origem, da mesma forma que na lente pontual.

```python
from scipy.optimize import brentq
from astropy.cosmology import FlatLambdaCDM

class sis_lens(object):
    """
    Esfera Isotérmica Singular
    """
    
    def __init__(self, cosmo, zl = 0.3, zs = 2.0, sigmav = 200):
        """
        Inicializamos o objeto SIS
        """
        
        self.sigmav = sigmav                      # Dispersão de velocidades
        self.cosmo = cosmo                        # Modelo Cosmológico utilizado
        self.zl = zl                              # Redshift da lente
        self.zs = zs                              # Redshift da Fonte
        
        
        # Calcula as distâncias diâmetro angular:
        self.dl = self.cosmo.angular_diameter_distance(self.zl)
        self.ds = self.cosmo.angular_diameter_distance(self.zs)
        self.dls = self.cosmo.angular_diameter_distance_z1z2(self.zl, self.zs)
        
        
        # Calcula o Raio de Einstein em arcsec
        self.thetaE = np.rad2deg( (4.0 * np.pi * sigmav**2 / (c /1000)**2 * self.dls/self.ds).value) * 3600.0

    
    def kappa(self, x):
        """
        Convergência
        """
        return 1 / (2*x)
    
    
    def gamma(self, x, phi):
        """
        Componentes do cisalhamento
        """
        return self.kappa(x) * np.cos(2*phi), self.kappa(x) * np.sin(2*phi)
    
    
    def mu(self, y1, y2):
        """
        Magnificação das duas imagens.
        Retorna respectivamente x+ e x- (caso tenha).
        """
        y = np.sqrt(y1**2 + y2**2)
        mu = []
        
        if y == 0:
            return 0
        
        else:
            mu.append(1 + 1/y)
        
        if y < 1:
            mu.append(abs(1 - 1/y))
            
        return mu
    
    
    def psi(self, x):
        """
        Potencial efetivo (obtido pela solução da EDO)
        """
        return np.log(x) + x
    
    
    def alpha(self, x):
        """
        Ângulo de deflexão
        """
        return x / abs(x)
    
    
    def cut(self):
        """
        Coordenadas da curva pseudo-cáustica
        """
        phi = np.linspace(0, 2*np.pi, 1000)
        
        y1 = np.cos(phi)
        y2 = np.sin(phi)
        return y1, y2
    
    
    def tan_caustic(self):
        """
        Coordenadas da curva cáustica tangencial. Para SIS é a origem.
        """
        return 0,0
    
    
    def tan_cc(self):
        """
        Coordenadas da curva crítica tangencial
        Para SIS é um círculo de raio 1
        """
        phi = np.linspace(0, 2*np.pi, 1000)
        
        x1 = np.cos(phi)
        x2 = np.sin(phi)
        return x1, x2
    
    
    def phi_ima(self, y1, y2):
        """
        Resolve a equação das lentes para uma dada posição da fonte
        """
        
        phi=[]
        xphi=[]
        
        y = np.sqrt((y1**2 + y2**2))             # Distância da fonte até origem
        
        # Calcula o ângulo polar para diversos casos
        if y == 0:
            ang = 0
            
        else:
            
            if y1 < 0 and y2 > 0:
                ang = np.pi - np.arcsin(y2 / y) 
                
            elif y1 < 0 and y2 < 0:
                ang = np.arcsin(abs(y2) / y) + np.pi
                
            elif y1 > 0 and y2 < 0:
                ang = 2 * np.pi - np.arcsin(abs(y2) / y) 
                
            else:
                ang = np.arcsin(y2 / y)
        
        
        # Fonte fora do Anel
        if y >= 1:
            return y + 1, ang             
        
        
        # Fonte dentro do Anel
        xphi.append(y + 1)
        phi.append(ang)
        
        xphi.append(1 - y)
        phi.append(ang + np.pi) 
        
        
        return xphi, phi
    
```

## 6 ) Elipse Isotérmica Singular
    
A Elipse ISotérmica Singular (SIE, no inglês) é um modelo semelhante ao da SIS e que também é muito utilizado, mas que possui resultados bem diferentes. A principal consequência de assumir uma lente com uma distribuição em forma elíptica é que isso remove a simetria axial, que simplificava as soluções. Da mesma forma que na SIS, vamos iniciar calculando as principais propriedades da lente.

A densidade superficial de massa agora será:
  
\begin{equation*}
                 \Sigma(\xi) = \frac{\sigma_v^2}{2g}\frac{\sqrt{f}}{\sqrt{\xi_1^2 + f^2\xi_2^2}}
\end{equation*}

onde $f$ é a razão entre os eixos da elipse e sempre estará na região $0 < f \leq 1$. Essa razão será uma das principais na generalização para o modelo de SIE. A partir da densidade superficial, temos que a convergência, em coordenadas adimensionais, será:

\begin{equation*}
            \kappa = \frac{\sqrt{f}}{2\sqrt{x_1^2 + f^2x_2^2}} 
\end{equation*}

Podemos também reescrever em coordenadas adimensionais polares ($x, \varphi$), que serão muito úteis para esse modelo de lente. Definindo $\Delta (\varphi) = \sqrt{\cos^2\varphi + f^2 \sin^2\varphi}$, temos

\begin{equation*}
            \kappa =  \frac{1}{2x}\frac{\sqrt{f}}{\Delta (\varphi)}
\end{equation*}

Podemos também obter o potencial efetivo. Escrevendo o laplacianos em coordenadas polares, teremos:

\begin{equation*}
            \frac{\partial^2 \psi}{\partial x^2} + \frac{1}{x} \frac{\partial \psi}{\partial x} + \frac{1}{x^2} \frac{\partial^2 \psi}{\partial \varphi^2} =\frac{\sqrt{f}}{2x \Delta (\varphi)}
\end{equation*}

Supondo uma separação de variáveis da seguinte forma: $\psi(x, \varphi) = x \tilde\psi(\varphi)$, e definindo $f' = \sqrt{1 - f^2}$, o resultado será:

\begin{equation*}
                \psi(x, \varphi) = x \frac{\sqrt{f}}{f'}\left( \sin (\varphi) \arcsin(f' \sin(\varphi))  + \cos(\varphi) \text{ arcsinh}(f' / f\cos(\varphi))\right )
\end{equation*}

Devido à quebra de simetria, o ângulo de deflexão terá duas componentes distintas, mas novamente ele não dependerá da distância até o centro da lente, somente do ângulo polar:

\begin{align*}
                \alpha_1(\varphi) &=  \frac{\sqrt{f}}{f'} \text{ arcsinh}\left( \frac{f'}{f} \cos(\varphi) \right) \\
                \alpha_2(\varphi) &= \frac{\sqrt{f}}{f'} \arcsin\left( f' \sin(\varphi) \right)
\end{align*}

O cisalhamento, da mesma forma que na SIS, será igual à convergência:

\begin{equation*}
                \gamma = \frac{\sqrt{f}}{2x \Delta (\varphi)}
\end{equation*}

Agora buscaremos as curvas críticas e as cáusticas do modelo SIE. Substituindo o resultado para a convergência e pro cisalhamento na equação dos autovalores da matriz jacobiana, temos que: $\lambda_t = 1 - 2\kappa$ e $\lambda_r = 1$. Ou seja, a SIE também não terá uma curva crítica radial e a magnificação também será somente tangencial. A curva crítica tangencial será uma elipse, dada pela equação
\begin{equation*}
                \vec x_t(\varphi) = \frac{\sqrt{f} }{\Delta (\varphi)}(\cos(\varphi), \sin(\varphi))
\end{equation*}

Agora que não temos mais simetria axial, a cáustica tangencial não será mais um ponto, mas sim uma curva extensa no plano das fontes.

\begin{align*}
                y_{t,1}(\varphi) &= \frac{\sqrt{f}}{\Delta (\varphi)} \cos(\varphi) - \frac{\sqrt{f}}{f'}\text{ arcsinh}\left(\frac{\sqrt{f}}{f'} \cos(\phi)  \right)\\
                 y_{t,2}(\varphi) &= \frac{\sqrt{f}}{\Delta (\varphi)} \sin(\varphi) - \frac{\sqrt{f}}{f'}\arcsin(f'\sin(\phi)  )
\end{align*}

Por fim, a SIE também terá uma curva de corte (*cut*), que dada pelo limite $y_{c} = \lim_{x\to 0} y(x, \varphi) =  -\alpha(\varphi)$. Ou seja, são os pontos no plano das fontes que são mapeados para a origem do plano da lente.

Importante mencionar também as regras para múltiplas imagens na SIE:
 - Se a fonte está fora da curva de corte e fora da cáustica, temos somente 1 imagem.
 - Se a fonte está dentro da curva de corte, o número de imagens aumenta em 1.
 - Se a fonte está dentro da cáustica, o número de imagens aumenta em 2.
 
Portanto, para lentes específicas de modelo SIE, é possivel ter 1, 2, 3 ou 4 imagens da mesma fonte.

### Exemplo e gráficos

Iremos definir uma nova classe para o modelo SIE e também adicionar um método numérico apresentado no livro do Meneghetti para encontrar todas as imagens de uma fonte.

```python
class sie_lens(object):
    """
    Elipse Isotérmica Singular
    """
    
    def __init__(self, cosmo, zl = 0.3, zs = 2.0, sigmav = 200, f = 0.6, pa = 45.0):
        """
        Inicializamos o objeto SIE
        """
        self.sigmav = sigmav                      # Dispersão de velocidades
        self.cosmo = cosmo                        # Modelo Cosmológico utilizado
        self.zl = zl                              # Redshift da lente
        self.zs = zs                              # Redshift da Fonte
        self.f=f                                  # Razão entre eixos
        self.pa=pa*np.pi/180.0                    # Ângulo de posição
        
        
        # Calcula as distâncias diâmetro angular:
        self.dl = self.cosmo.angular_diameter_distance(self.zl)
        self.ds = self.cosmo.angular_diameter_distance(self.zs)
        self.dls = self.cosmo.angular_diameter_distance_z1z2(self.zl, self.zs)
        
        # Calcula o Raio de Einstein em arcsec
        self.thetaE = np.rad2deg( (4.0 * np.pi * sigmav**2 / (c /1000)**2 * self.dls/self.ds).value) * 3600.0

        
    def delta(self,f,phi):
        return np.sqrt( np.cos(phi - self.pa)**2 + self.f**2 * np.sin(phi-self.pa)**2 )

    
    def kappa(self,x,phi):
        """
        Convergênca em função da coordenada polar
        """
        return(np.sqrt(self.f)/ (2.0 * x * self.delta(self.f,phi)) )

    
    def gamma(self,x,phi):
        """
        Componentes do Cisalhamento em função da coordenada polar
        """
        return(-self.kappa(x, phi) * np.cos( 2.0*phi - self.pa ),
        -self.kappa(x, phi) * np.sin( 2.0*phi - self.pa) )
    
    
    def mu(self, x, phi):
        """
        Magnificação em função da coordenada polar
        """
        ga1, ga2 = self.gamma(x, phi)
        ga = np.sqrt(ga1*ga1 + ga2*ga2)
        return 1.0 / ( (1.0 - self.kappa(x,phi) - ga) * (1.0 - self.kappa(x,phi) + ga) )

    def psi_til(self, phi):
        """
        Parte angular do potencial efetivo
        """
        if (self.f < 1.0):
            fp = np.sqrt(1.0 - self.f**2)
            a = np.sin(phi - self.pa) * np.arcsin(fp*np.sin(phi - self.pa))
            b = np.cos(phi - self.pa) * np.arcsinh(fp*np.cos(phi - self.pa)/ f)
            
            return np.sqrt(self.f) * (a + b) / fp
        
        else:
            return(1.0)
        
    
    def psi(self,x,phi):
        """
        Potencial efetivo completo
        """
        return x * self.psi_til(phi)
    
    
    def alpha(self,phi):
        """
         Ângulo de deflexão em termos do ângulo polar
        """
        fp = np.sqrt(1.0 - self.f**2)
        
        a1 = np.sqrt(self.f) * np.arcsinh(fp * np.cos(phi)/self.f) / fp 
        a2 = np.sqrt(self.f) * np.arcsin(fp * np.sin(phi)) / fp
        return a1, a2
    
    def cut(self):
        """
        Coordenadas da curva de corte
        """
        phi = np.linspace(0, 2*np.pi, 1000)
        
        y1_,y2_ = self.alpha(phi)
        y1 = y1_ * np.cos(self.pa) - y2_ * np.sin(self.pa)
        y2 = y1_ * np.sin(self.pa) + y2_ * np.cos(self.pa)
        return -y1,-y2
    
    
    def tan_caustic(self):
        """
        Coordenadas da curva cáustica tangencial
        """
        phi = np.linspace(0, 2*np.pi, 1000)
        delta = np.sqrt( np.cos(phi)**2 + self.f**2 * np.sin(phi)**2 )
        a1, a2 = self.alpha(phi)
        
        y1_=np.sqrt(self.f)/delta*np.cos(phi)-a1
        y2_=np.sqrt(self.f)/delta*np.sin(phi)-a2
        
        y1 = y1_ * np.cos(self.pa) - y2_ * np.sin(self.pa)
        y2 = y1_ * np.sin(self.pa) + y2_ * np.cos(self.pa)
        return y1,y2
    
    
    def tan_cc(self):
        """
        Coordenadas da curva crítica tangencial
        """
        phi = np.linspace(0, 2*np.pi, 1000)
        delta = np.sqrt( np.cos(phi)**2 + self.f**2 * np.sin(phi)**2 )
        r = np.sqrt(self.f) / delta
        
        x1 = r * np.cos(phi + self.pa)
        x2 = r * np.sin(phi + self.pa)
        return(x1,x2)
    
    
    def x_ima(self,y1,y2,phi):
        """
        Distância da image para o centro da lente
        """
        x = y1*np.cos(phi) + y2*np.sin(phi) + (self.psi_til(phi+self.pa))
        return x


    def phi_ima(self,y1,y2,checkplot=True,eps=0.001,nphi=100):
        """
        Método numérico para solução da equação da lente.
        """
        
        y1_ = y1 * np.cos(self.pa) + y2 * np.sin(self.pa)
        y2_ = - y1 * np.sin(self.pa) + y2 * np.cos(self.pa)

        
        # Função obtida da equação da lente
        # O método se resume a encontrar os zero dessa função
        def phi_func(phi):
            a1, a2 = self.alpha(phi)
            func = (y1_+ a1) * np.sin(phi) - (y2_ + a2) * np.cos(phi)
            return func

        
        # Calculo os valores e o sinal da função acima 
        U = np.linspace(0., 2.0*np.pi + eps , nphi)
        c = phi_func(U)
        s = np.sign(c)
        
        phi = []
        xphi = []
        
        #Realiza um loop pelos ângulos polares
        for i in range(len(U)-1):
            
           # Se dois valores em sequência tem sinais opostos,
           # usa o método brentq() para encontrar o ponto
            if s[i] + s[i+1] == 0: 
                u = brentq(phi_func, U[i], U[i+1])
                z = phi_func(u)

                if np.isnan(z) or abs(z) > 1e-3:      # Caso dê erro, pula para prox iteração
                    continue

                    
                x = self.x_ima(y1_, y2_, u)
                
                # Adiciona a solução para a lista  se a distância
                # radial é positiva
                if (x > 0):
                    phi.append(u)
                    xphi.append(x)

                    
        xphi = np.array(xphi)
        phi = np.array(phi)

        # Retorna as coordenadas polares das imagens
        # Adiciona ângulo de posição para voltar ao plano rotacionado da lente
        return xphi, phi + self.pa
```

Ou seja, os pontos dentro da curva de corte no primeiro quadrante geram uma imagem adicional no terceiro quadrante. Para verificarmos que esse é realmente o caso, vamos realizar mais um teste com fontes espalhadas.

```python
# Configurações para produção do gráfico
fig,ax=plt.subplots(1,2,figsize=(10,5))
ms = 6
col = ['g', 'r', 'c', 'm', 'y' , 'brown', 'grey', 'pink']

ax[0].set_title("Plano das Fontes", fontsize=14)
ax[0].axhline(0, linestyle='--', alpha=0.3)
ax[0].axvline(0, linestyle='--', alpha=0.3)
ax[0].set_ylim(-2, 2)
ax[0].set_xlim(-2, 2)
ax[0].set_xlabel("y1", fontsize=14)
ax[0].set_ylabel("y2", fontsize=14)
ax[0].plot(y1_cau, y2_cau, color='k', markersize = ms)              # Cáustica Tangencial
ax[0].plot(y1_cut, y2_cut, color='b')                               # Curva de Corte



ax[1].set_title("Plano da Lente", fontsize=14)
ax[1].axhline(0, linestyle='--', alpha=0.3)
ax[1].axvline(0, linestyle='--', alpha=0.3)
ax[1].set_ylim(-2, 2)
ax[1].set_xlim(-2, 2)
ax[1].set_xlabel("x1", fontsize=14)
ax[1].set_ylabel("x2", fontsize=14)
ax[1].plot(x1_cc, x2_cc, color="k")                  # Curva Crítica Tangecial


img = [[-0.1, 0.1], [-0.1, -0.2], [0.6, 0.3], [0.3, -0.5], [-0.4, -0.4]]
       

for i in range(0, len(img)):
    y1, y2 = img[i][0], img[i][1]
    
    # Posição e magnificação das imagens
    x, phi = sie.phi_ima(y1, y2)
    ms_ima = abs( sie.mu(x, phi) )
    
    for j in range(len(ms_ima)):
        ms_ima[j] *= 30

    # Convertemos para coord. cartesianas
    x1_ima = x * np.cos(phi)
    x2_ima = x * np.sin(phi)
    
    # Adicionamos fontes e imagens ao gráfico
    ax[0].plot(y1, y2, 'o', ms, color=col[i])  
    ax[1].scatter(x1_ima, x2_ima, ms_ima, color=col[i])
```

Comparando a posição das fontes e das imagens, podemos verificar que esse é realmente o padrão. Agora iremos testar também o caso de produzir 3 imagens. Isso ocorre quando a fração $f$ é suficientemente pequena para que a cáustica se extenda além da curva de corte, dessa forma, é possível que uma fonte esteja dentro da cáustica, mas forma da curva de corte.

```python
sigmav = 200.0
f = 0.15
pa = 0.0
co = FlatLambdaCDM(Om0=0.3,H0=70.0)
sie = sie_lens(co,sigmav=sigmav,f=f,pa=pa)          # Inicializa o objeto SIE


# Calcula as coordenadas da curva de corte,
# curva crítica e da cáustica
y1_cut, y2_cut = sie.cut()
y1_cau, y2_cau = sie.tan_caustic()
x1_cc, x2_cc = sie.tan_cc()


# Configurações para produção do gráfico
fig,ax=plt.subplots(1,2,figsize=(10,5))
ms = 6
col = ['g', 'r', 'c', 'm', 'y' , 'brown', 'grey', 'pink']
lim = 3

ax[0].set_title("Plano das Fontes", fontsize=14)
ax[0].axhline(0, linestyle='--', alpha=0.3)
ax[0].axvline(0, linestyle='--', alpha=0.3)
ax[0].set_ylim(-lim, lim)
ax[0].set_xlim(-lim, lim)
ax[0].set_xlabel("y1", fontsize=14)
ax[0].set_ylabel("y2", fontsize=14)
ax[0].plot(y1_cau, y2_cau, color='k', markersize = ms)              # Cáustica Tangencial
ax[0].plot(y1_cut, y2_cut, color='b')                               # Curva de Corte



ax[1].set_title("Plano da Lente", fontsize=14)
ax[1].axhline(0, linestyle='--', alpha=0.3)
ax[1].axvline(0, linestyle='--', alpha=0.3)
ax[1].set_ylim(-lim, lim)
ax[1].set_xlim(-lim, lim)
ax[1].set_xlabel("x1", fontsize=14)
ax[1].set_ylabel("x2", fontsize=14)
ax[1].plot(x1_cc, x2_cc, color="k")                  # Curva Crítica Tangecial


img = [[0., 1.6], [0., 1.45], [0., 1.3], [0., 1.15],
       [0., 0.9], [0., 0.45], [0., 0.75], [0., 0.6]]

for i in range(0, len(img)):
    y1, y2 = img[i][0], img[i][1]
    
    # Posição e magnificação das imagens
    x, phi = sie.phi_ima(y1, y2)
    ms_ima = abs( sie.mu(x, phi) )
    
    for j in range(len(ms_ima)):
        ms_ima[j] *= 50
    
    # Convertemos para coord. cartesianas
    x1_ima = x * np.cos(phi)
    x2_ima = x * np.sin(phi)
    
    # Adicionamos fontes e imagens ao gráfico
    ax[0].plot(y1, y2, 'o', ms, color=col[i])  
    ax[1].scatter(x1_ima, x2_ima, ms_ima, color=col[i])
```
